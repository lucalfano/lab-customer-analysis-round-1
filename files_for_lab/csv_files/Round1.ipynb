```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
file1 = file1.rename(columns={'st':'state'})
file2 = file2.rename(columns={'st':'state'})
```

```python
cols = []
for column in file1.columns:
    cols.append(column.lower())
file1.columns = cols
file1
```

```python
cols = []
for column in file2.columns:
    cols.append(column.lower())
file2.columns = cols
file2
```

```python
cols = []
for column in file3.columns:
    cols.append(column.lower())
file3.columns = cols
file3
```

```python
# file2 = file2[file1.columns]
# file3 = file3[file1.columns]

# display.file1(head)
# display.file2(head)
# display.file3(head)
file2 = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file3 = file3 [['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
```

```python
# concecate
dataset = pd.concat([file1, file2, file3], axis=0)
dataset
```
